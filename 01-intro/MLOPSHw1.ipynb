```python
df['duration'] = pd.to_datetime(df['tpep_dropoff_datetime']) - pd.to_datetime(df['tpep_pickup_datetime'])
```

```python
df.head()
```

```python
df['duration'] = df['duration'].dt.total_seconds() / 60
```

Ans 2 - 42.59 mins

```python
df_out = df.loc[(df['duration'] >= 1) & (df['duration'] <= 60)]
```

```python
df_out.head()
```

```python
df_out.shape[0]/df.shape[0]
```

Ans 3 - 98.12 %

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
df_out['PULocationID'] = df_out['PULocationID'].astype(str)
df_out['DOLocationID'] = df_out['DOLocationID'].astype(str)
```

```python
df_out.head()
```

```python
data_dicts = df_out[['PULocationID','DOLocationID']].to_dict(orient='records')
```

```python
vec = DictVectorizer(sparse=True)
feature_matrix = vec.fit_transform(data_dicts)
```

```python
print(feature_matrix.shape)
```

Ans 4 - 515

```python
# We now use the above to predict the duration of the ride.
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = feature_matrix
Y = df_out['duration']
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
y_pred = model.predict(X)

rmse = np.sqrt(mean_squared_error(Y, y_pred))
print("RMSE", rmse)
```
